```python

import re
from collections import Counter

def count_keywords(transcripts, keywords):
    keyword_counts = {keyword: 0 for keyword in keywords}
    for transcript in transcripts:
        for keyword in keywords:
            # Use regex to find whole words only, case insensitive
            keyword_counts[keyword] += len(re.findall(r'\b' + re.escape(keyword) + r'\b', transcript, re.IGNORECASE))
    return keyword_counts

keywords = [
    'Voluntary  Cancel', 'Booking', 'IRROPS', 'Upgrade', 'Seating',
    'Mileage   Plus', 'Checkout', 'Voluntary   Change', 'Post Flight',
    'Check In', 'Other  Topics', 'Communications', 'Schedule Change',
    'Products & Services', 'IRROPS  ', 'Digital   Support',
    'Seating  ', 'Disability', 'Unaccompanied Minor', ' Baggage',
    'Traveler  Updates', 'Communications  ', 'ETC', 'Upgrade  ',
    'Unaccompanied Minor  ', 'Voluntary Change', 'Voluntary Change  ',
    'Checkout  ', 'Mileage Plus', 'Mileage Plus  ', 'Booking  ',
    'Baggage  ', 'Post-Flight', 'Post-Flight  ', 'Schedule Change  ',
    'Baggage', 'Traveler Updates', 'Voluntary Cancel', 'Check-In',
    'Products and Services', 'Check-In  ', 'Other Topics',
    'Other Topics  ', 'ETC  ', 'Disability  ', 'Digital Support',
    'Digital Support  ', 'Voluntary Cancel  ',
    'Products and Services  ', 'Traveler Updates  ',
    'Traveler   Updates', 'Digital  Support', 'Mileage  Plus',
    'Voluntary  Change'
]

# Count occurrences of keywords in call transcripts
keyword_counts = count_keywords(callsdata['call_transcript'], keywords)

# Display the results
print("Keyword occurrences in call transcripts:")
for keyword, count in keyword_counts.items():
    print(f"{keyword}: {count}")
```

```python
# Define key terms to look for in the call transcripts
key_terms = ['cancel', 'reschedule', 'upgrade', 'seating']

# Create a function to count occurrences of each term in transcripts
def count_keyword_occurrences(transcripts, keywords):
    keyword_counts = {key: 0 for key in keywords}
    for transcript in transcripts:
        for key in keywords:
            if key in transcript.lower():
                keyword_counts[key] += 1
    return keyword_counts

# Apply the function to the call transcripts
keyword_counts = count_keyword_occurrences(callsdata['call_transcript'], key_terms)

# Plotting the keyword occurrences
plt.figure(figsize=(10, 6))
sns.barplot(x=list(keyword_counts.keys()), y=list(keyword_counts.values()), palette='coolwarm')
plt.title('Occurrences of Self-Solvable Keywords in Call Transcripts')
plt.xlabel('Keyword')
plt.ylabel('Number of Occurrences')
plt.show()
```
